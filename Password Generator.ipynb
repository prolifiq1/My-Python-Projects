```python
import random
import string

def generate_password(num_letters, num_symbols, num_numbers):
    """Generate a random password with specified character types."""
    password_chars = []

    # Add random letters
    for _ in range(num_letters):
        password_chars.append(random.choice(string.ascii_letters))

    # Add random symbols
    for _ in range(num_symbols):
        password_chars.append(random.choice(string.punctuation))

    # Add random numbers
    for _ in range(num_numbers):
        password_chars.append(random.choice(string.digits))

    # Shuffle to mix the characters
    random.shuffle(password_chars)

    # Join into a string
    return ''.join(password_chars)

def get_positive_integer(prompt):
    """Get a valid positive integer from user input."""
    while True:
        try:
            value = int(input(prompt))
            if value < 0:
                print("Please enter a non-negative number.")
                continue
            return value
        except ValueError:
            print("Please enter a valid number.")

def main():
    print("=== Password Generator ===")

    # Get user preferences
    num_letters = get_positive_integer("How many letters would you like? ")
    num_symbols = get_positive_integer("How many symbols would you like? ")
    num_numbers = get_positive_integer("How many numbers would you like? ")

    # Check if password will be empty
    total_chars = num_letters + num_symbols + num_numbers
    if total_chars == 0:
        print("\nYour password would be empty! Please choose at least one character.")
        return

    # Generate and display password
    password = generate_password(num_letters, num_symbols, num_numbers)
    print(f'\nYour password is: {password}')
    print(f'\nPassword length: {total_chars} characters.')

if __name__ == "__main__":
    main()
```
