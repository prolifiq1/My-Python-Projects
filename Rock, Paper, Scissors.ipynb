```python
import random

rock = """
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
"""

paper = """
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
"""

scissors = """
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
"""

choices = [rock, paper, scissors]
choice_names = ['Rock', 'Paper', 'Scissors']

while True:
    try:
        user_choice = int(input('What do you want to choose? Type 0 for Rock, 1 for Paper or 2 for Scissors'))
        if user_choice in [0, 1, 2]:
            break
        else:
            print('Please enter 0, 1, or 2 only.')
    except ValueError:
        print('That is not a number! Please enter 0, 1, or 2 only.')

computer_choice = random.randint(0, 2)

print(f'You chose {choice_names[user_choice]}: {choices[user_choice]}')
print(f'Computer chose {choice_names[computer_choice]}: {choices[computer_choice]}')

if user_choice == computer_choice:
    print('It is a draw!')
elif (user_choice == 0 and computer_choice == 2) or \
     (user_choice == 1 and computer_choice == 0) or \
     ( user_choice == 2 and computer_choice == 1):
    print('You win! üèÜ')
else:
    print('Computer wins! üòî')
```
