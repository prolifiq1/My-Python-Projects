### Checking if the player has won

```python
import random

word_list = ['apple', 'banana', 'cherry']
chosen_word = random.choice(word_list)
print(chosen_word)

# Create blanks
display = []
for letter in chosen_word:
    display.append('_')

# Game loop - keep playing until they win
game_over = False

while not game_over:
    guess = input('Guess a letter: ').lower()

    # Replace blanks with correct guesses
    for position in range(len(chosen_word)):
        letter = chosen_word[position]
        if letter == guess:
            display[position] = letter

    print(display)

    # Check if player has won
    if '_' not in display:
        game_over = True
        print('You win! ü•≥')
```

### Keeping track of the player's lives

```python
import random

word_list = ['apple', 'banana', 'cherry']
chosen_word = random.choice(word_list)
print(chosen_word)

# Create blanks
display = []
for letter in chosen_word:
    display.append('_')

# Game setup
game_over = False
lives = 6    # Player starts with 6 lives

while not game_over:
    guess = input('Guess a letter: ').lower()

    # Replace blanks with correct guesses
    for position in range(len(chosen_word)):
        letter = chosen_word[position]
        if letter == guess:
            display[position] = letter

    # Check if guess was wrong
    if guess not in chosen_word:
        lives -= 1     # Lose a life
        print(f'Wrong! You have {lives} lives left.')

        if lives == 0:
            game_over = True
            print('You lose! üíÄ')

    print(display)

    # Check if player has won
    if '_' not in display:
        game_over = True
        print('You won. üèÜ')
```
